```python
import pandas as pd
from pathlib import Path
```

```python
# File
results_path = Path("../data/results")
all_files = list(results_path.glob("utmb_*.csv"))

dfs = []
for file in all_files:
    year = int(file.stem.split("_")[1])
    df = pd.read_csv(file)
    df["year"] = year
    # Remplacer runnerUri par runner_id si pr√©sent
    if "runnerUri" in df.columns:
        df = df.rename(columns={"runnerUri": "runner_id"})
    dfs.append(df)

df_all = pd.concat(dfs, ignore_index=True)
print(df_all.shape)
df_all.head()
```
